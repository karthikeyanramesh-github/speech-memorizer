```python
# user input for audio data to convert text data using google recognition with english or tamil text
# import libraries
import pyaudio
import wave
import threading
import time
import datetime  
from os import path
from scipy.io.wavfile import write
from scipy.io.wavfile import read
import os
import speech_recognition as sr
import sqlite3
#enter a filename for audio data
name =input("Enter your file name : ")
# Record in chunks of 1024 samples
chunk = 1024 
# 16 bits per sample
sample_format = pyaudio.paInt16
channels = 1
# Record at 44100 samples per second
fs = 44100
# Record for 15 seconds
seconds = 15 
# Initialize PyAudio
p = pyaudio.PyAudio()
# Start Recording
stream = p.open(format=sample_format,
                channels=channels,
                rate=fs,
                frames_per_buffer=chunk,
                input=True)
# Initialize array to store frames
frames = [] 
# Audio Recording using pyaudio library
print("Recording...")
for i in range(0, int(fs / chunk * seconds)):
    data = stream.read(chunk)
    frames.append(data)
# Stop Recording
stream.stop_stream()
stream.close()
p.terminate()
print("Finished Recording")
# Save the recorded audio
# Get the current time
now = datetime.datetime.now() 
# Create a unique filename
filename = name + ".wav" 
wavefile = wave.open(filename, 'wb')
wavefile.setnchannels(channels)
wavefile.setsampwidth(p.get_sample_size(sample_format))
wavefile.setframerate(fs)
wavefile.writeframes(b''.join(frames))
wavefile.close()
print(filename)
path = filename
file_path = os.path.abspath(path)
print(file_path)
recognizer = sr.Recognizer()
audio_ex = sr.AudioFile(file_path)
type(audio_ex)
# Create audio data
with audio_ex as source:
    audiodata = recognizer.record(audio_ex)
type(audiodata)
# Extract text using speech recognition
eng_text = recognizer.recognize_google(audio_data=audiodata, language='en-US')
tam_text = recognizer.recognize_google(audio_data=audiodata, language='ta-IN')
hin_text = recognizer.recognize_google(audio_data=audiodata, language='hi-IN')
tel_text = recognizer.recognize_google(audio_data=audiodata, language='te-IN')
mal_text = recognizer.recognize_google(audio_data=audiodata, language='ml-IN')
now=datetime.datetime.now()
print("ENGLISH TEXT : "+eng_text)
print(now)
print("TAMIL TEXT : "+tam_text)
print(now)
print("HINDI TEXT : "+hin_text)
print(now)
print("TELUGU TEXT : "+tel_text)
print(now)
print("MALAYALAM TEXT : "+mal_text)
print(now)
print("Saved audio file as: " + filename)
# Get the absolute path of the audio file
path = os.path.abspath(filename)
print("Audio file path: " + path)
```

```python
# Split keyword for searching audio data to help retrieve in database
# import libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
# example text data
sen=eng_text 
# tokenize the text into words
words = word_tokenize(eng_text.lower())
# remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]
# extract keywords based on part-of-speech tagging
pos_tags = nltk.pos_tag(filtered_words)
keywords = [word for word, pos in pos_tags if pos.startswith('NN')]
# store keywords in an array
keyword_array = []
for word in keywords:
    if word not in keyword_array:
        keyword_array.append(word)
print(keyword_array)
```
